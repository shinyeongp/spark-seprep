```python
# Hypothesis Formulation
```

Hypothesis 1: There is a positive correlation between fixed acidity and density of red wine.

Hypothesis 2: The citric acid content of white wine is higher than that of red wine.

Null: citric acid content of red wine = citric acid content of white wine.

Alternative: citric acid content of red wine < citric acid content of white wine.

```python
# Hypothesis Testing
```

```python
# Hypothesis Test 1
from scipy.stats.stats import pearsonr
x = df_redwine['density']
y = df_redwine['fixed acidity']
pearsonr(x, y) # finds the correlation coefficient (r) and the p-value
```

```python
# Documentation and Reporting for hypothesis test 1
```

The correlation coefficient shows that there is a strong, positive relationship between density and fixed acidity. The correlation coefficient is close to 1. Since the p-value is less than 0.05, we have statistically significant evidence that there is a correlation between density and fixed acidity. 

```python
# Hypothesis Test 2
from scipy.stats import ttest_ind 

sample_red = df_redwine['citric acid']
sample_white = df_whitewine['citric acid']
t_stat, p_value = ttest_ind(sample_red, sample_white) # finds t statistic and p value

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
# Documentation and Reporting for hypothesis test 2
```

Since p-value is significanly less than a = 0.05, we can reject the null hypothesis. We have sufficient evidence for the alternative that the citric acid content of white wine is higher than that of red wine.
