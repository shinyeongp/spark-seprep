```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/Users/neri/Downloads/BostonUniversity/Fall2023/DS219/bank+marketing/bank/bank.csv",
                 sep=';', header=0, quotechar='"', encoding='utf-8')
```

```python
# Exploratory Data Analysis (EDA)
# Explore the dataset's structure and summary statistics

print("First 5 rows of dataset:")
print(df.head())

print("\nDataset info:")
print(df.info())

print("\nSummary statistics:")
print(df.describe())

print("\n Data Columns:")
print(df.columns)
```

```python
# Visualize key features using plots

# distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.plot()

# distribution of the job title
plt.figure(figsize=(12, 6))
sns.countplot(x='job', data=df)
plt.title('Distribution of Jobs')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# distribution of the education levels
plt.figure(figsize=(10, 6))
sns.countplot(x='education', data=df)
plt.title('Distribution of Education Levels')
plt.xlabel('Education')
plt.ylabel('Count')
plt.show()
```

```python
# Hypothesis Formulation
# Formulate at least two hypotheses related to the dataset
 
# Hypothesis 1: 
# Individuals who may be older, may have a higher balance.

# Hypothesis 2: 
# Individuals with housing loans may be more or less likely to have personal loans as well. 
```

```python
# Hypothesis Testing
# Test the formulated hypotheses using statistical tests or visualizations
# Hypothesis 1:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='balance', data=df)
plt.title('Scatter Plot of Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()

# Hypothesis 2:
plt.figure(figsize=(8, 6))
sns.countplot(x='housing', hue='loan', data=df)
plt.title('Loan Status by Housing')
plt.xlabel('Those w Housing Loans')
plt.ylabel('Count')
plt.show()
```
