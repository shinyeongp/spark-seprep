## White Wine

```python
# Importing dataset 
url = '/Users/moonieboo/Desktop/BU/DS 219/Assignment 5/wine+quality/winequality-white.csv'
df_whiteW = pd.read_csv(url, sep=';')
df_whiteW
```

```python
# Display the first few rows of the dataset
print(df_whiteW.head())

# Summary statistics
print(df_whiteW.describe())

# Check for missing values
print(df_whiteW.isnull().sum())

# Distribution of wine quality
sns.countplot(x='quality', data=df_whiteW)
plt.title('Distribution of Wine Quality')
plt.show()
```

## Hypothesis
1. the wine quality of White Wine is higher than Red Wine
2. Acidity levels impact the quality of the wine

```python
# for the first hypothesis
# null: wine quality of White Wine = wine quality of Red Wine 
# Alternative: White Wine's quality is higher than Red Wine's quality 


from scipy.stats import ttest_ind

sample_red = df_redW['quality']
sample_white = df_whiteW['quality']
t_test, p_value = ttest_ind(sample_red, sample_white)

print("t_test value: ", t_test)
print("p_value: ", p_value)
```

##### Since p_value is less than 0.05, we can say that this hypothesis is significant as this have enough evidence. Therefore, the wine quality of white wine is higher than Red wine. 

```python
# for the second hypothesis
# Boxplot for acidity levels vs. red wine quality
sns.boxplot(x='quality', y='fixed acidity', data=df_redW)
plt.title('Acidity Levels vs. Red Wine Quality')
plt.show()

# Boxplot for acidity levels vs. White wine quality
sns.boxplot(x='quality', y='fixed acidity', data=df_whiteW)
plt.title('Acidity Levels vs. White Wine Quality')
plt.show()
```

```python
## Comparing a specific Quality level with overall mean fixed acidity in Red Wine 
from scipy import stats

# Select a specific quality level
quality_level = 5
group = df_redW[df_redW['quality'] == quality_level]['fixed acidity']

# T-test
t_stat, p_val = stats.ttest_1samp(group, df_redW['fixed acidity'].mean())

print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
## Comparing a specific Quality level with overall mean fixed acidity in White Wine 
from scipy import stats

# Select a specific quality level
quality_level = 5
group = df_whiteW[df_whiteW['quality'] == quality_level]['fixed acidity']

# T-test
t_stat, p_val = stats.ttest_1samp(group, df_whiteW['fixed acidity'].mean())

print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

#### Since the P-values for both Red Wine and White Wine are less than 0.05 by showing the values as 0.011(round up) and 0.00032(round up), we can reject the null hypothesis. In this context, the null hypothesis would be that there is no difference in the mean fixed acidity levels between the groups being compared. Rejecting it suggests that there is a significant difference.
